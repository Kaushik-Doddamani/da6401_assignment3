```python
import pandas as pd
# from ace_tools import display_dataframe_to_user

# Load the model predictions
preds = pd.read_csv('../predictions_vanilla/predictions.csv')

# Build character-level pairs of (actual, predicted)
pairs = []
for actual_str, pred_str in zip(preds['target'], preds['prediction']):
    # Align up to the shorter length to avoid index errors
    for a_char, p_char in zip(actual_str, pred_str):
        pairs.append((a_char, p_char))

# Create a DataFrame and compute the confusion matrix via crosstab
pairs_df = pd.DataFrame(pairs, columns=['Actual', 'Predicted'])
confusion_matrix = pd.crosstab(
    pairs_df['Actual'],
    pairs_df['Predicted'],
    rownames=['Actual'],
    colnames=['Predicted'],
    dropna=False
)

from IPython.display import display
# Display the confusion matrix to the user
display('confusion_matrix', confusion_matrix)
```
